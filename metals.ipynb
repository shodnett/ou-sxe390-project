```python
# import python packages
import pandas as pd
import numpy as np
from scipy import stats as sci
import time as t
from matplotlib import pyplot as plt


# importing macrotidal (high tidal range) and microtidal (low tidal range) datasets
high = pd.read_excel("Appendix_4.xlsx", 'High', engine = 'openpyxl')
low = pd.read_excel('Appendix_4.xlsx', 'Low', engine = 'openpyxl')


# allocating variables to lists for macrotidal estuaries, sediment then dissolved
hsed_cd = high['Sed_Cd'].dropna().tolist()

hsed_pb = high['Sed_Pb'].dropna().tolist()

hsed_hg = high['Sed_Hg'].dropna().tolist()

hdis_cd = high['Diss_Cd'].dropna().tolist()

hdis_pb = high['Diss_Pb'].dropna().tolist()

hdis_hg = high['Diss_Hg'].dropna().tolist()


# allocating variables to lists for microtidal estuaries, sediment then dissolved
lsed_cd = low['Sed_Cd'].dropna().tolist()

lsed_pb = low['Sed_Pb'].dropna().tolist()

lsed_hg = low['Sed_Hg'].dropna().tolist()

ldis_cd = low['Diss_Cd'].dropna().tolist()

ldis_pb = low['Diss_Pb'].dropna().tolist()

ldis_hg = low['Diss_Hg'].dropna().tolist()

```

```python
# Hypothesis 1 - There is a difference in sediment cadmium between macrotidal and microtidal UK estuaries 
# mann-whitney u test on macrotidal and microtidal medians
h7 = sci.mannwhitneyu(hsed_cd, lsed_cd)
print(h7)
# produces p-value and U statistic


# plot box and whisker plot and take medians of the two groups
fig7 = plt.figure()
hsed_cdmed = np.median(hsed_cd)
hsed_cdmedlab = round(hsed_cdmed, 3)
lsed_cdmed = np.median(lsed_cd)
lsed_cdmedlab = round(lsed_cdmed, 3)
yaxis = 'Macrotidal', 'Microtidal'

ax7 = fig7.add_subplot()
h7x = [hsed_cd, lsed_cd]

h7plot = plt.boxplot(h7x, tick_labels=yaxis)
for line in h7plot['medians']:
    line.set_color('orange')
    line.set_linewidth(3)
ax7.set_ylabel('Sediment Cd (mg $kg^{-1}$)')
ax7.set_title('Sediment cadmium in macrotidal and microtidal estuaries', fontsize=12, fontweight='bold', wrap=True)
plt.text(1.1, 0.253, hsed_cdmedlab)
plt.text(2.1, 0.442, lsed_cdmedlab)
plt.show()
```

```python
# Hypothesis 2 - There is a difference in sediment lead between macrotidal and microtidal UK estuaries
h8 = sci.mannwhitneyu(lsed_pb, hsed_pb)
print(h8)

fig8 = plt.figure()
hsed_pbmed = np.median(hsed_pb)
hsed_pbmedlab = round(hsed_pbmed, 3)
lsed_pbmed = np.median(lsed_pb)
lsed_pbmedlab = round(lsed_pbmed, 3)

ax8 = fig8.add_subplot()
h8x = [hsed_pb, lsed_pb]

h8plot = plt.boxplot(h8x, tick_labels=yaxis)
for line in h8plot['medians']:
    line.set_color('orange')
    line.set_linewidth(3)
ax8.set_ylabel('Sediment Pb (mg $kg^{-1}$)')
ax8.set_title('Sediment lead in macrotidal and microtidal estuaries', fontsize=12, fontweight='bold', wrap=True)
plt.text(1.1, 63.28, hsed_pbmedlab)
plt.text(2.1, 22.41, lsed_pbmedlab)
plt.show()
```

```python
# Hypothesis 4 - There is a difference in dissolved cadmium between macrotidal and microtidal UK estuaries
h10 = sci.mannwhitneyu(hdis_cd, ldis_cd)
print(h10)

fig10 = plt.figure()
hdis_cdmed = np.median(hdis_cd)
hdis_cdmedlab = round(hdis_cdmed, 3)
ldis_cdmed = np.median(ldis_cd)
ldis_cdmedlab = round(ldis_cdmed, 3)

ax10 = fig10.add_subplot()
h10x = [hdis_cd, ldis_cd]

h10plot = plt.boxplot(h10x, tick_labels=yaxis)
for line in h10plot['medians']:
    line.set_color('orange')
    line.set_linewidth(3)
ax10.set_ylabel('Dissolved Cd (µg $L^{-1}$)')
ax10.set_title('Dissolved cadmium in macrotidal and microtidal estuaries', fontsize=12, fontweight='bold', wrap=True)
plt.text(1.1, 0.043, hdis_cdmedlab)
plt.text(2.1, 0.05, ldis_cdmedlab)
plt.show()
```
